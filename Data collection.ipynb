```python
from google.cloud import bigquery
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "pd-prod-bristol-e0be-6c1ab6198c00.json"
def query():
    
    client = bigquery.Client()
    query_job = client.query(
     """SELECT *
 FROM `pd-prod-bristol-e0be.looker_connectors.historical_rts2_enriched` 
 WHERE DATE(PUBLICATION_DATE_AND_TIME) = CURRENT_DATE()-1 

 """ 
     )
    results = query_job.result() # Waits for job to complete. 
    destination = query_job.destination
    destination = client.get_table(destination)
    print("The query data:")
    done = False
    start_at_record = 0
    page_size = 20
    while not done:
        rows = client.list_rows(destination, page_size=20, 
        start_index=start_at_record)
        count = 0
        for row in rows:
            count +=1
            print(count, " trading_date_time={}, price={},quantity={}".format(row["TRADING_DATE_AND_TIME"], row["PRICE"], row["QUANTITY"]))
        if count != page_size:
            done = True
        start_at_record += page_size
    
query()
```
