```python
from google.cloud import bigquery
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "pd-prod-bristol-e0be-6c1ab6198c00.json"
def query():
    
    client = bigquery.Client()
    query_job = client.query(
     """SELECT *

FROM `pd-prod-bristol-e0be.looker_connectors.historical_rts2_enriched` 

LIMIT 100
     """ 
     )
    results = query_job.result() # Waits for job to complete. 
    print(results)
    count = 0 
    for row in results:       
        print(count, " Trade date time {} : Price {}".format(row.TRADING_DATE_AND_TIME, row.PRICE))
    count +=1
query()
```
